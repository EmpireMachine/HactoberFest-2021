```python
import pandas as pd
import numpy as np
data = pd.read_csv(r"C:\Users\Dell\Desktop\ML Lab\MVRdata.csv")
data
```

```python
train_data = data[ :80]
train_data_X = train_data[["x0", "x1", "x2"]].to_numpy()
train_data_Y = train_data[["y"]].to_numpy()
train_data_X
train_data_Y

#Here training data preparation is happening
#1st 80 rows are used
```

```python
test_data = data[80: ]
test_data_X = test_data[["x0", "x1", "x2"]].to_numpy()
test_data_Y = test_data[["y"]].to_numpy()
test_data_X
test_data_Y

#Here testing data preparation is happening
#Remaining data after 80 Rows are used
```

```python
Xt = np.transpose(train_data_X)
inv_of_Xt_x_X = np.linalg.inv(np.dot(Xt, train_data_X))
Weight_matrix = np.dot(inv_of_Xt_x_X, np.dot(Xt, train_data_Y))

#Weight matrix is calculated using pseudo inverse
```

```python
Y_pred = np.dot(test_data_X, Weight_matrix)
test_data = test_data.drop(columns="x0")
test_data["Y_pred"] = Y_pred
test_data
absolute_error = squared_error = 0

#Bias column from test data is dropped and then prediction happens
```

```python
num_test_data = len(test_data)
for index, row in test_data.iterrows():
    error = row["y"] - row["Y_pred"]
    absolute_error += abs(error)
    squared_error += error * error
    
#After prediction, squared error is calculated for each point
```

```python

mean_absolute_error = absolute_error / num_test_data
mean_squared_error = squared_error / num_test_data
```

```python
print("Mean Absolute Error: ", mean_absolute_error)
print("Mean Squared Error: ", mean_squared_error)
```
